```python
# # Calculate accuracy

def calculate_accuracy(tv, dp):
    count = 0
    total = len(dp)
    for i in range(total):
        if(tv[i] == dp[i]):
            count = count + 1

    return count / total
```

```python
calculate_accuracy(true_values, data_prediction)
```
